```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filtered_data = pd.read_csv("Filtered_Trade_Data__Russia___China_.csv")
```

```python
filtered_data
```

```python
# Get unique years for x-ticks
years = filtered_data["Year"].dropna().unique()

china_data = filtered_data[filtered_data["Commercial Partner"] == "China"]
russia_data = filtered_data[filtered_data["Commercial Partner"] == "Russia"]

# China Plot
plt.figure(figsize=(12, 6))
plt.plot(china_data["Year"], china_data["Import Value (CHF)"], label="China - Import Value (CHF)", marker="o")
plt.plot(china_data["Year"], china_data["Export Value (CHF)"], label="China - Export Value (CHF)", marker="o")
plt.title("Trade Values with China (CHF)")
plt.xlabel("Year")
plt.ylabel("Value (CHF)")
plt.xticks(years)  # Set x-ticks to be each year
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Russia Plot
plt.figure(figsize=(12, 6))
plt.plot(russia_data["Year"], russia_data["Import Value (CHF)"], label="Russia - Import Value (CHF)", marker="o")
plt.plot(russia_data["Year"], russia_data["Export Value (CHF)"], label="Russia - Export Value (CHF)", marker="o")
plt.title("Trade Values with Russia (CHF)")
plt.xlabel("Year")
plt.ylabel("Value (CHF)")
plt.xticks(years)  # Set x-ticks to be each year
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
